```python
import pandas as pd
```

```python
df = pd.read_csv('/content/data.csv')
```

```python
df
```

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer, TfidfVectorizer
```

```python
text = df['content']
```

```python
text
```

```python
tfidf = TfidfVectorizer()
tfidf_matrix = tfidf.fit_transform(text)
print(tfidf_matrix)
```

# Interpretasi
Format "(i, j) x.y" mewakili bahwa dalam dokumen i, istilah j memiliki skor TF-IDF sebesar x.y.

(0, 3251) 0.1055299758616986: Pada dokumen pertama (dokumen 0), istilah di indeks 3251 memiliki skor TF-IDF sekitar 0.1055.
dan seterusnya untuk istilah lain dalam dokumen 0 dan dokumen berikutnya.

Output ini menunjukkan skor TF-IDF untuk semua istilah dalam semua dokumen.

Skor TF-IDF yang lebih tinggi menunjukkan bahwa sebuah istilah penting dalam dokumen tertentu, dan sering digunakan dalam  pengambilan informasi .

```python
tfidf.get_feature_names_out()
```

# Interpretasi
array teresebut adalah fitur-fitur/kata-kata yang terdapat dalam dokumen tersebut.

```python
pd.DataFrame(tfidf_matrix.toarray(), columns=tfidf.get_feature_names_out(), index=text)
```

# Interpretasi
Angka-angka tiap kolom yang kebanyakan bernilai 0 adalah akibat tidak adanya representasi kata tersebut pada kalimat yang ditampilkan. Kolom yang dihasilkan juga terlalu banyak karena pada kasus ini data belum dibersihkan sehingga banyak noise seperti '00','000', '0sebelumnya', dll.
Seharusnya jika kita dapat melihat semua fitur/kolom yang ada maka kata-kata yang muncul pada kalimat pertama akan memiliki nilai pada kolom kata tersebut, nilainya didasarkan dengan rumus TF-IDF yaitu:
**TF-IDF(t, d) = TF(t, d) * IDF(t)**,
dengan

**TF(t, d) = (jumlah kemunculan kata t dalam dokumen d) / (jumlah kata total dalam dokumen d)**,
dan
**IDF(t) = log(N / (jumlah dokumen yang mengandung kata t))**
