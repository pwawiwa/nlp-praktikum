# 3d. Lakukan stopword removing berdasarkan hasil 3c, simpan hasilnya dalam
bentuk csv/excel. (bobot:10)

```python
import Sastrawi
```

```python
slang_words
```

```python
teks = slang_matches
```

```python
from Sastrawi.StopWordRemover.StopWordRemoverFactory  import StopWordRemoverFactory
import re
def remove_stopword(words):
  factory = StopWordRemoverFactory()
  stopword_remover = factory.create_stop_word_remover()
  processed_words = [stopword_remover.remove(word) for word in words]
  return processed_words
```

```python
processed_texts = remove_stopword(teks)
```

```python
processed_texts
```

```python
clean_processed_texts= [item for item in processed_texts if item != '']

print(clean_processed_texts)
```

# 3.e Lakukan stemming berdasarkan hasil 3d dan tampilkan 100 stem pertama.
(bobot:10)


```python
#3.e Stemming

from Sastrawi.Stemmer.StemmerFactory import StemmerFactory

# Buat objek stemmer
factory = StemmerFactory()
stemmer = factory.create_stemmer()

# Kata yang akan di-stem
kata = clean_processed_texts

stemmed_kata = [stemmer.stem(kata) for kata in kata]

# Output kata-kata yang telah di-stem
print(stemmed_kata)
```
